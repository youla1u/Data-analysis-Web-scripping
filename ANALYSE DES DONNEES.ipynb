```python

############################################################################################################################## 
############################### Analysons  les nombres des praticiens par commune ############################################ 
```

```python
# Ici nous calculons le nombre de pratien pour chaque commune de la Loire Atlantique

grouped_df = Donnees_extraites.groupby("COMMUNE")

grouped_df = grouped_df.agg({"Nom Prenom": "nunique"})

grouped_df = grouped_df.reset_index()

grouped_df.rename(columns = {'Nom Prenom':'Nb_pratient_par_com'}, inplace = True)

grouped_df  
```

```python
# Voyons à l'aide d'un histogramme comment ce nombre de praticien par commune est distribué  

hist = grouped_df['Nb_pratient_par_com'].hist()
```

```python
# L'histogramme ci-dessus indique que les communes où il y a un nombre faible de praticien sont les plus nombreux. 
# Voyons si il existe des communes exceptionnelles(qui pouraient avoir un nombre très élevé de praticien par exemple).   

sns.boxplot(x=grouped_df['Nb_pratient_par_com'])
plt.show
```

```python
# La boite à moustache indique bien la présence des valeurs aberrantes. 

# Dans la suite de notre analyse nous allons donc supprimer ces valeurs aberrantes 
# afin connaitre le nombre moyen de praticien par commune.Car la moyenne est sensible 
# aux valeurs aberrantes(outliers)  
```

```python
# Ici nous avons les différentes statistiques, sans le retrait de valeurs aberrantes.
# le nombre moyen (mean) de praticien par commune vaut aproximativement 5 (mean=4.936000) 

grouped_df['Nb_pratient_par_com'].describe()
```

```python
# Détection et suppression des valeurs abérentes: 

# Nous utiliserons la règle de règle IQR pour détecter les valeurs aberrantes. 
# Tout d'abord, nous calculerons l'intervalle interquartile des données (IQR = Q3 — Q1).
# Puis, nous déterminerons les limites de valeurs aberrantes avec IQR.

# Nous obtiendrons la limite inférieure avec ce calcul Q1–0.25*IQR. 
# Nous obtiendrons la limite supérieure avec ce calcul Q3 + 0.45*IQR.

# Selon cette règle, les données entre les limites sont acceptables mais les données 
# en dehore des limites iférieure et supérieure sont des valeurs aberrantes. 
```

```python

Q1=grouped_df['Nb_pratient_par_com'].quantile(0.25) #premier quartile (Q1),valeur au-dessous de laquelle se trouvent 25 % des données 
                                                    #lorsqu'elles sont arrangées en ordre croissant. 
Q3=grouped_df['Nb_pratient_par_com'].quantile(0.75) #troisième quartile (Q3),valeur au-dessous de laquelle se trouvent 75 % des données 
                                                    #arrangées en ordre croissant
IQR=Q3-Q1 # intervalle interquartile des données (IQR = Q3 — Q1)
```

```python
inf_lim=Q1-0.25*IQR #limite inférieure
sup_lim=Q3+0.45*IQR #la limite supérieure
print('inf_lim:',inf_lim,'  sup_lim:',sup_lim)
```

```python
outliers_inf_1=(grouped_df['Nb_pratient_par_com']<inf_lim)# données en dehore de la  limite inférieure.
outliers_sup_1=(grouped_df['Nb_pratient_par_com']>sup_lim) # données en dehore de la  limite supérieur.
```

```python
#Suppression des outliers (données en dehore des limites iférieure et supérieure )
grouped_df_2=grouped_df[~(outliers_inf_1|outliers_sup_1)]
```

```python
# histogramme sans outliers
hist = grouped_df_2['Nb_pratient_par_com'].hist()
```

```python
# Après le retait des valeurs aberrantes nous voyons clairement que les communes avec un seul praticien sont 
# plus nombreux. les commune avec 7  praticiens sont les moins nombreus.
# Aussi on voit une décroissance génerale du nombre de commune en fonction du nombre de pratitien  
```

```python
# La boite à moustache ci-dessous montre clairement que les valeurs aberrantes ont été retiréés.
sns.boxplot(x=grouped_df_2['Nb_pratient_par_com'])
plt.show
```

```python
# On peut voir que le nombre moyen de praticien par commune est 3 (mean 3.116505) aulieu de 5 obtenu précédement(présence 
# des valeurs abérentes).
# le minimum est 1 et le maximum est 7  
grouped_df_2['Nb_pratient_par_com'].describe()
```

```python


###############################################################################################################################
###########################################  Analyse des données population/praticien #########################################
```

```python
# Ici nous ajoutons la colone "Nombre de population par praticien" à la base de données extraite (depuis ameli.fr)

# Pour une commune donnée la population/pratitien = à la population totale de cette commune divisé par le nombre de 
# praticien dans cette comune. On arrondira le résultat pour avoir un nombre entier  

N=grouped_df.shape[0]
M=Donnees_extraites.shape[0]
Donnees_extraites["Population_par_praticien"]=""
for i in range(N):
    for j in range(M):
        if Donnees_extraites['COMMUNE'][j]==grouped_df['COMMUNE'][i]:
            Donnees_extraites["Population_par_praticien"][j]=round(Donnees_extraites['PTOT'][j]/grouped_df['Nb_pratient_par_com'][i])    
```

```python
Donnees_extraites['Population_par_praticien'] =Donnees_extraites['Population_par_praticien'].astype(int)# conversion en nombre entier
Donnees_extraites
```

```python
# histogramme sans retrait des valeurs aberrantes de la colonne 'Population_par_praticien'.   
hist = Donnees_extraites['Population_par_praticien'].hist()
```

```python
#Comme l'histogramme ci-dessus est dificile à interpréter nous allons donc vérifier à l'aide la boite à moustache ci-dessous 
# s'il y a des valeurs aberrantes dans la colonne 'Population_par_praticien'.

data =Donnees_extraites
sns.boxplot(x=data['Population_par_praticien'])
plt.show

```

```python
# La boite à moustache indique bien la présence des outliers dans les données de 'Population_par_praticien'. 
```

```python
# La moyenne provisoire est de "3587 individus/praticien".   
data['Population_par_praticien'].describe()
```

```python
# Extraction de valeurs abérantes. 
Q1=data['Population_par_praticien'].quantile(0.25)
Q3=data['Population_par_praticien'].quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
inf_lim=Q1-0.68*IQR
sup_lim=Q3+1.05*IQR
print('inf_lim:',sup_lim,'  sup_lim:',sup_lim)
```

```python
outliers_inf_2=(data['Population_par_praticien']<inf_lim)
outliers_sup_2=(data['Population_par_praticien']>sup_lim)
```

```python
# retrait des outliers 
df_2=data[~(outliers_inf_2|outliers_sup_2)]
```

```python
# Importation des packages 
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
# url pour la collecte des données de la Loire Atlantique
url="http://annuairesante.ameli.fr/trouver-un-professionnel-de-sante/chirurgiens-dentistes/44-loire-atlantique"
```

```python
reponse=requests.get(url)
soup=BeautifulSoup(reponse.text,"html.parser")
print(soup.prettify()) # contenu html de l'url
```

```python
D=soup.findAll(id="site_inner")[0]
D
```

```python
# Puisque sur la page précedente il y a 3 listes de communes de la loire Atlantique 
# nous allons les traiter séparement
First=D.find_all("ul", class_="first")[0] # première liste
Second=D.find_all("ul", class_="second")[0]# deuxième liste
Third=D.find_all("ul", class_="third")[0]# troisième liste
```

```python
# initialisation de trois listes pour les remplire des urls des communes  
L1=[]
L2=[]
L3=[]
```

```python
# L1 Contiendra les premiers urls des communes
for f in First.findAll("li"):
    for link in f.find_all('a'):
        L1.append("http://annuairesante.ameli.fr"+link.get('href'))

# L2 Contiendra les deuxièmes urls des communes
for f in Second.findAll("li"):
    for link in f.find_all('a'):
        L2.append("http://annuairesante.ameli.fr"+link.get('href')) 

# L3 Contiendra les troisièmes urls des communes        
for f in Third.findAll("li"):
    for link in f.find_all('a'):
        L3.append("http://annuairesante.ameli.fr"+link.get('href')) 
        
        
```

```python
# Nous allons exécuter ce code pour récupérer les urls de tous les praticiens des la première liste L1 communes.
# Après avoir récolté toute les cordonnées de ces praticiens en suivant les codes suivant, nous allons le réexécuter 
# pour la liste L2 puis pour la liste L3 en le modifiant légèrement. 
ll=[]
for url_2 in L1: # L2 L3 . Modification à faire
    reponse_2=requests.get(url_2)
    soup_2=BeautifulSoup(reponse_2.text,"html.parser")
    D_2=soup_2.find_all("div", class_="liste-professionnel")[0]
    for d in D_2.find_all("div", class_="item-professionnel"):   
        dd = d.find_all("div", class_="item-professionnel-inner")[0]
        link = dd.find_all('a')[0]
        ll.append("http://annuairesante.ameli.fr"+link.get('href'))
ll        
```

```python
len(ll)# nombre d'url des praticiens
```

```python
# Cette fonction va nous permettre de récupérer les éléments de l'adresse d'un praticien 

from bs4 import NavigableString
from bs4 import Tag
def adress(v):
    text = ''
    for child in v:
        if isinstance(child, NavigableString):
            text += str(child).strip()
        elif isinstance(child, Tag):
            if child.name != 'br':
                text += child.text.strip()
            else:
                text += '\n'

    result = text.strip().split('\n')
    return result
```

```python
lll=[]
```

```python
# Récupération de toutes les information dont on a besoin pour les praticien 
for i in range(len(ll)):
    dic_i={}
    link=ll[i]
    reponse=requests.get(link)
    soup=BeautifulSoup(reponse.text,"html.parser")
    v=soup.find_all("div", class_="nom_pictos")[0]  
    dic_i["Nom Prénom"]=v.text.replace("\n","").replace("\n\n","")
    v2=""
    if soup.find_all("div", class_="item left adresse")==[]:
        v2=soup.find_all("h2", class_="item left adresse")[0]
    elif soup.find_all("h2", class_="item left adresse")==[]:
         v2=soup.find_all("div", class_="item left adresse")[0]
    if len(adress(v2))==3:
        dic_i["SOCIETE"]=adress(v2)[0]
        dic_i["BATIMENT"]="None"
        dic_i["ADRESSE"]=adress(v2)[1]
        dic_i["Coplement ADR"]="None"
        dic_i["COMMUNE"]=adress(v2)[2]
    elif len(adress(v2))==4:
        dic_i["SOCIETE"]=adress(v2)[0]
        dic_i["BATIMENT"]=adress(v2)[1]
        dic_i["ADRESSE"]=adress(v2)[2]
        dic_i["Coplement ADR"]="None"
        dic_i["COMMUNE"]=adress(v2)[3] 
    elif len(adress(v2))==5:
        dic_i["SOCIETE"]=adress(v2)[0]
        dic_i["BATIMENT"]=adress(v2)[1]
        dic_i["ADRESSE"]=adress(v2)[2]
        dic_i["Coplement ADR"]=adress(v2)[3]
        dic_i["COMMUNE"]=adress(v2)[4]  
    elif len(adress(v2))==2:
        dic_i["SOCIETE"]="None"
        dic_i["BATIMENT"]="None"
        dic_i["ADRESSE"]=adress(v2)[0]
        dic_i["Coplement ADR"]="None"
        dic_i["COMMUNE"]=adress(v2)[1]
    print(i)
    lll.append(dic_i)
```
